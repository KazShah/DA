```python
path=r'C:/Users/kazim/Documents/Master of Data/2021/SIT782 Project part B/Powercor/'
#path=(r'C:\Users\kazim\Documents\Master of Data\2021\SIT782 Project part B\Combine')
#path = os.getcwd()
files = os.listdir(path)
File_count=0
data_power_temp_xlb=[]
files_xlsx_xb = [f for f in files if f[14:19] == 'reads']

for f in files_xlsx_xb:
    df_xb=pd.read_excel(path+f,engine='pyxlsb')
    data_power_temp_xlb.append(df_xb)
    print(f)
    File_count=File_count+1 
data_powercore_xb=pd.concat(data_power_temp_xlb)
print("Total Files Loaded: ",File_count)
```

```python
data_powercore_xb = data_powercore_xb.dropna(how='any',axis=0)
data_powercore_xlsx=data_powercore_xlsx.dropna(how='any',axis=0)
```

```python
# Change column names
data_powercore_xb.rename(columns={'channel_type':'channel'}, inplace=True)
```

```python
#Standardising data types
data_powercore_xb['read_date'] = pd.to_datetime(data_powercore_xb['read_date'])
data_powercore_xlsx['read_date'] = pd.to_datetime(data_powercore_xlsx['read_date'])
```

```python
data_powercore_complete=pd.concat([data_powercore_xlsx,data_powercore_xb],ignore_index=True, sort=False)
data_powercore_complete.nmi=data_powercore_complete.nmi.astype(int)
```

```python
#Mapping nmi, postcodes and city values to powercore data
#Reading nmi master file
path=r'C:/Users/kazim/Documents/Master of Data/2021/SIT782 Project part B/Powercor/'
files = os.listdir(path)
files_nmi = [f for f in files if f == 'c4net request id.xlsb']
for f in files_nmi:
    df_xb=pd.read_excel(path+f,engine='pyxlsb')
data_nmi=pd.DataFrame(df_xb)
```

```python
data_power_master = pd.merge(data_powercore_complete,data_nmi,on ='nmi',how ='inner')
```
